### 状態空間モデル ベイズ型統計モデル

```python
# 推移行列などの初期化
def FGHset(n_dim_trend, n_dim_obs=1, n_dim_series=0, Q_sigma2=10):
    n_dim_Q = (n_dim_trend!=0) + (n_dim_series!=0)
    if n_dim_series>0:
        n_dim_state = n_dim_trend + n_dim_series - 1
    else:
        n_dim_state = n_dim_trend
        
    # 各種行列の初期化
    G = np.zeros((n_dim_state, n_dim_Q))
    F = np.zeros((n_dim_state, n_dim_state))
    H = np.zeros((n_dim_obs, n_dim_state))
    Q = np.eye(n_dim_Q) * Q_sigma2
    
    # 各行列のトレンド成分に対するブロック行列を構築
    G[0, 0] = 1
    H[0, 0] = 1
    
    # トレンドモデルの推移行列の構築
    if n_dim_trend == 1:
        F[0, 0] = 1
    elif n_dim_trend == 2:
        F[0, 0] = 2
        F[0, 1] = -1
        F[1, 0] = 1
    elif n_dim_trend == 3:
        F[0, 0] = 3
        F[0, 1] = -3
        F[0, 2] = 1
        F[1, 0] = 1
        F[2, 1] = 1
        
    # PyKalman の Q は G.dot(Q).dot(G.T) を想定しているため Q を update
    Q = G.dot(Q).dot(G.T)
    
    return n_dim_state, F, H, Q
```

```python
# 観測値の次元数
n_dim_obs = 1

# トレンドの次元数
n_dim_trend = 2

# 推移行列などの定義
n_dim_state, F, H, Q = FGHset(n_dim_trend, n_dim_obs)
```

```python
# 状態の平均値ベクトルの初期値
initial_state_mean = np.zeros(n_dim_state)

# 状態の分散共分散行列の初期値
initial_state_covariance = np.ones((n_dim_state, n_dim_state))

# カルマンフィルタのモデル生成
kf = KalmanFilter(
    # 1：観測値の次元数
    n_dim_obs = n_dim_obs,
    # k：状態の次元数
    n_dim_state=n_dim_state,
    # x_0：状態の平均値ベクトルの初期値（k次元）
    initial_state_mean=initial_state_mean,
    # V_0：状態の分散共分散行列の初期値（k x k次元）
    initial_state_covariance=initial_state_covariance,
    # F：推移行列（k x k 次元）
    transition_matrices=F,
    # H：観測行列（1 x k 次元）
    observation_matrices=H,
    # R：観測ノイズ w の分散共分散行列（1 x 1次元。観測値が1次元の場合はスカラ）
    observation_covariance=1.0,
    # Q：システムノイズ v の分散共分散行列
    transition_covariance=Q
)
```

```python
# フィルタ
filtered_state_means, filtered_state_covs = kf.filter(y_train["y"])
# フィルタ＋平準化
smoothed_state_means, smoothed_state_covs = kf.smooth(y_train["y"])
```

```python
pred_o_smoothed = smoothed_state_means.dot(H.T)
pred_o_smoothed.shape
```

```python
# 長期予測格納用のベクトルを用意（値はすべて入れ替わるため 0 で初期化しなくてよい）
pred_y = np.empty(len(y_test["y"]))

# 現在の状態と分散共分散行列を取得
current_state = smoothed_state_means[-1]
current_cov = smoothed_state_covs[-1]
for i in range(len(y_test["y"])):
    # filter_update は観測値を入力しなければ１期先予測のみを実行する
    current_state, current_cov = kf.filter_update(current_state,
                                                 current_cov,
                                                 observation=None)
    pred_y[i] = kf.observation_matrices.dot(current_state)
```

```python
fig = plt.figure(figsize=(24, 8))

plt.plot(y_train["ds"][-731:-1], y_train["y"][-731:-1], label="y_train")
plt.plot(y_test["ds"], y_test["y"], label="y_test")
plt.plot(y_test["ds"], pred_y, label="long_pred")

plt.title("Tokyo tempreture 1990-2019")
plt.xlabel("time")
plt.ylabel("tempreture")

plt.legend()
plt.show()
```

# 季節調整モデル

```python
# 推移行列などの初期化
def FGHset(n_dim_trend, n_dim_obs=1, d_dim_series=0, Q_sigma2=10):
    n_dim_Q = (n_dim_trend != 0) + (n_dim_series != 0)
    if n_dim_series > 0:
        n_dim_state = n_dim_trend + n_dim_series - 1
    else:
        n_dim_state = n_dim_trend
        
    # 各種行列の初期化
    G = np.zeros((n_dim_state, n_dim_Q))
    F = np.zeros((n_dim_state, n_dim_state))
    H = np.zeros((n_dim_obs, n_dim_state))
    Q = np.eye(n_dim_Q) * Q_sigma2
    
    # トレンドモデルのブロック行列の構築
    G[0, 0] = 1
    H[0, 0] = 1
    if n_dim_trend == 1:
        F[0, 0] = 1
    elif n_dim_trend == 2:
        F[0, 0] = 2
        F[0, 1] = -1
        F[1, 0] = 1
    elif n_dim_trend == 3:
        F[0, 0] = 3
        F[0, 1] = -3
        F[0, 2] = 1
        F[1, 0] = 1
        F[2, 1] = 1
        
    start_elem = n_dim_trend
    start_col = n_dim_trend
    
    # 季節調整成分のブロック行列の構築
    if n_dim_series > 0:
        G[start_elem, 1] = 1
        H[0, start_elem] = 1
        for i in range(n_dim_series-1):
            F[start_elem, start_elem + i] = -1
        for i in range(n_dim_series-2):
            F[start_elem+i+1, start_elem + i] = 1
            
    # PyKalman の Q は G.dot(Q).dot(G.T) を想定しているため Q を update
    Q = G.dot(Q).dot(G.T)
    
    return n_dim_state, F, H, Q
```

```python
# 観測値の次元数
n_dim_obs = 1
# トレンドの次元数
n_dim_trend = 2
# 季節成分の次元数
n_dim_series = 12

# 推移行列などの定義
n_dim_state, F, H, Q = FGHset(n_dim_trend, n_dim_obs, n_dim_series)
```

```python
# 状態の平均値ベクトルの初期値
initial_state_mean = np.zeros(n_dim_state)
# 状態の分散共分散行列の初期値
initial_state_covariance = np.ones((n_dim_state, n_dim_state))

# カルマンフィルタのモデル生成
kf = KalmanFilter(
    # 1：観測値の次元数
    n_dim_obs = n_dim_obs,
    # k：状態の次元数
    n_dim_state=n_dim_state,
    # x_0：状態の平均値ベクトルの初期値（k次元）
    initial_state_mean=initial_state_mean,
    # V_0：状態の分散共分散行列の初期値（k x k次元）
    initial_state_covariance=initial_state_covariance,
    # F：推移行列（k x k 次元）
    transition_matrices=F,
    # H：観測行列（1 x k 次元）
    observation_matrices=H,
    # R：観測ノイズ w の分散共分散行列（1 x 1次元。観測値が1次元の場合はスカラ）
    observation_covariance=1.0,
    # Q：システムノイズ v の分散共分散行列(m x m 次元)
    transition_covariance=Q
)
```

```python
# フィルタ + 平準化
smoothed_state_means, smoothed_state_covs = kf.smooth(y_train["y"])
pred_o_smoothed = smoothed_state_means.dot(H.T)
```

```python
pred_y = np.empty(len(y_test["y"]))
current_state = smoothed_state_means[-1]
current_cov = smoothed_state_covs[-1]
for i in range(len(y_test["y"])):
    current_state, current_cov = kf.filter_update(current_state,
                                                 current_cov,
                                                 observation=None)
    pred_y[i] = kf.observation_matrices.dot(current_state)
```

```python
fig = plt.figure(figsize=(24, 8))

plt.plot(y_train["ds"][-731:-1], y_train["y"][-731:-1], label="y_train")
plt.plot(y_test["ds"], y_test["y"], label="y_test")
plt.plot(y_test["ds"], pred_y, label="seasonal_adjusted_long_pred")

plt.title("Tokyo tempreture 1990-2019")
plt.xlabel("time")
plt.ylabel("tempreture")

plt.legend()
plt.show()
```

### ハイパーパラメタ最適化

```python
# 繰り返し回数は10回
# ハイパーパラメタ更新の対象は F, H, Q, R
emed_kf = kf.em(y_train["y"], n_iter=10, em_vars='all')
```

```python
# ハイパーパラメタ F、H、Q、R 更新後の平準化系列
em_smoothed_state_means, em_smoothed_state_covs = emed_kf.smooth(y_train["y"])

# 階差
diff = 2
```

```python
# トレンド成分
# トレンド成分では階差の次数分の状態が対象
t_em_pred_o_smoothed = np.dot(em_smoothed_state_means[:, :diff],
                             emed_kf.observation_matrices[:, :diff].T)

# 季節成分
# 季節成分は状態ベクトルにおいてトレンド成分移行の要素が対象
s_em_pred_o_smoothed = np.dot(em_smoothed_state_means[:, diff:],
                             emed_kf.observation_matrices[:, diff:].T)

current_state = em_smoothed_state_means[-1]
current_cov = em_smoothed_state_covs[-1]

# トレンド成分と季節成分の長期予測値を格納するベクトルを用意
t_pred_y = np.empty(len(y_test["y"]))
s_pred_y = np.empty(len(y_test["y"]))

# トレンド成分及び季節成分の長期予測
for i in range(len(y_test["y"])):
    current_state, current_cov = emed_kf.filter_update(current_state,
                                                      current_cov,
                                                      observation=None)
    t_pred_y[i] = \
        emed_kf.observation_matrices[:, :diff].dot(current_state[:diff])
    s_pred_y[i] = \
        emed_kf.observation_matrices[:, diff:].dot(current_state[diff:])
```

```python
fig = plt.figure(figsize=(24, 8))

plt.plot(np.hstack([t_em_pred_o_smoothed.flatten(), t_pred_y]),
        '--', label="trend")
plt.plot(np.hstack([s_em_pred_o_smoothed.flatten(), s_pred_y]),
        ':', label="seasonal")
plt.plot(
    np.hstack([
        s_em_pred_o_smoothed.flatten() + t_em_pred_o_smoothed.flatten(),
        s_pred_y + t_pred_y]),
    '+-', label="forecast"
)

plt.title("Tokyo tempreture 1990-2019")
plt.xlabel("time")
plt.ylabel("tempreture")

plt.legend()
plt.show()
```
